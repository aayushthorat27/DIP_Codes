```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/Edge detectn.bmp',cv2.IMREAD_GRAYSCALE)
print(img.shape)
```

```python
plt.figure(figsize = (10,5))
plt.title("Original Image")
plt.axis("off")
plt.imshow(img,cmap = 'gray')
plt.show()
```

```python
image_resized = cv2.resize(img,(128,128))
plt.figure(figsize = (10,5))
plt.title("Resized Image 128*128")
plt.axis("off")
plt.imshow(image_resized,cmap = 'gray')
plt.show()
```

```python
hist = cv2.calcHist([img],channels = [0],mask =None,histSize=[256],ranges = [0,256])
plt.figure(figsize = (12,5))
plt.title("Histogram of the Image")
plt.xlabel("Pixel value")
plt.ylabel("Frequency")
plt.axis("on")
plt.plot(hist)
plt.show()
```

```python
# rotating the image
rotated_image90 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
rotated_image90_counter = cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotated_image180 = cv2.rotate(img,cv2.ROTATE_180)
plt.figure(figsize = (12,5))
plt.subplot(1,4,1)
plt.title("90 Clockwise")
plt.axis("off")
plt.imshow(rotated_image90,cmap = 'gray')

plt.subplot(1,4,2)
plt.title("90 Counter Clockwise")
plt.axis("off")
plt.imshow(rotated_image90_counter,cmap = 'gray')

plt.subplot(1,4,3)
plt.title("180 Rotated")
plt.axis("off")
plt.imshow(rotated_image180,cmap = 'gray')


plt.subplot(1,4,4)
plt.title("Original Image")
plt.axis("off")
plt.imshow(img,cmap = 'gray')
plt.show()

```

```python
# Thresholding of the Image
_,threshold_image = cv2.threshold(img,128,255,cv2.THRESH_BINARY)
plt.subplot(1,2,1)
plt.title("Original Image")
plt.axis("off")
plt.imshow(img,cmap = 'gray')

plt.subplot(1,2,2)
plt.title("Binary threshold Image")
plt.axis("off")
plt.imshow(threshold_image,cmap = 'gray')
plt.show()
```

```python
# Adding Noise to the image
noise  = np.zeros_like(img,dtype = np.float32)
cv2.randn(noise , 0, 100)
noisy_image = cv2.add(img.astype(np.float32),noise)
noisy_image = np.clip(noisy_image,0,255).astype(np.uint8)

plt.title("Noisy Image")
plt.axis("off")
plt.imshow(noisy_image,cmap = 'gray')
plt.show()
```
