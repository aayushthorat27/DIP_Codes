# PERFORM EDGE DETECTION USING VARIOUS MASKS
1. Robert's operator
2. Prewitt operator
3. Sobel operator

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("DIP3E_Original_Images_CH10/Fig1016(a)(building_original).tif",cv2.IMREAD_GRAYSCALE)
plt.title("Original Image")
plt.axis("off")
plt.imshow(img,cmap = 'gray')
plt.show()
```

```python
# Convolution from scratch
def conv2D(image, kernel):
    ih, iw = image.shape
    kh, kw = kernel.shape  # Get kernel dimensions from the kernel
    pad_h = kh // 2
    pad_w = kw // 2
    padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')
    output_img = np.zeros_like(image, dtype=np.float32)
    for i in range(ih):
        for j in range(iw):
            region_of_interest = padded_img[i : i + kh, j : j + kw]
            output_img[i, j] = np.sum(region_of_interest * kernel)
    return output_img
```

1. Robert Masks

```python
robert_x = np.array([[-1,0],
                    [0,-1]],dtype = np.float32)
robert_y = np.array([[0,-1],
                    [-1,0]],dtype = np.float32)
r_x = conv2D(img,robert_x)
r_y = conv2D(img,robert_y)
robert_mag = np.sqrt(r_x**2 + r_y**2).astype(np.uint8)

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.axis("off")
plt.tight_layout()
plt.imshow(img,cmap = "gray")

plt.subplot(1,2,2)
plt.title("Robert Operator")
plt.axis("off")
plt.tight_layout()
plt.imshow(robert_mag,cmap = "gray")

plt.show()
```

2. Prewitt operator

```python
prewitt_x = np.array([[-1,-1,-1],
                    [0,0,0],
                     [1,1,1]],dtype = np.float32)
prewitt_y = np.array([[-1,0,1],
                    [-1,0,1],
                     [-1,0,1]],dtype = np.float32)
p_x = conv2D(img,prewitt_x)
p_y = conv2D(img,prewitt_y)
prewitt_mag = np.sqrt(p_x**2 + p_y**2).astype(np.uint8)

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.axis("off")
plt.tight_layout()
plt.imshow(img,cmap = "gray")

plt.subplot(1,2,2)
plt.title("Prewitt Operator")
plt.axis("off")
plt.tight_layout()
plt.imshow(prewitt_mag,cmap = "gray")

plt.show()
```

3. Sobel Operator

```python
sobel_x = np.array([[-1,-2,-1],
                    [0,0,0],
                     [1,2,1]],dtype = np.float32)
sobel_y = np.array([[-1,0,1],
                    [-2,0,2],
                     [-1,0,1]],dtype = np.float32)
s_x = conv2D(img,sobel_x)
s_y = conv2D(img,sobel_y)
sobel_mag = np.sqrt(p_x**2 + p_y**2).astype(np.uint8)

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.axis("off")
plt.tight_layout()
plt.imshow(img,cmap = "gray")

plt.subplot(1,2,2)
plt.title("Sobel Operator")
plt.axis("off")
plt.tight_layout()
plt.imshow(sobel_mag,cmap = "gray")

plt.show()
```
